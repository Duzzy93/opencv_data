```python
import matplotlib.pyplot as plt
import cv2

img = cv2.imread("images/matplot.jpg")
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)


plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(6, 6))
plt.imshow(img_rgb)
plt.title("  - / ")
plt.axis('on')
plt.show()
```
